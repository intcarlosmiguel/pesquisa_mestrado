```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
dir = [i for i in os.listdir('./C/output/infect/') if('vacinado' not in i)]
```

```python
dir
```

```python
vac = []
for i in dir:
    x = np.loadtxt(f'./C/output/infect/{i}')
    x = np.array([i for i in x if(sum(np.isnan(i)) == 0)]).T
    vac.append(np.vstack((np.arange(0,x.shape[1]/2,0.5),x)))
```

```python
for infect,file in zip(vac,dir):
    #plt.plot(infect[0],infect[1]/2029,label = f'Suscetíveis {file[9:-4]}')
    plt.plot(infect[0],infect[2]/2029,label = f'Expostos {file[9:-4]}')
plt.legend()
plt.grid()
plt.ylim(0.01,0.1)
plt.xlim(50,300)
plt.show()
```

```python

for infect,file in zip(vac,dir):
    #plt.plot(infect[0],infect[1]/2029,label = f'Suscetíveis {file[9:-4]}')
    plt.plot(infect[0],infect[-1]/2029,label = f'Mortos {file[9:-4]}')
plt.legend()
plt.grid()
plt.show()
```

```python

for infect,file in zip(vac,dir):
    #plt.plot(infect[0],infect[1]/2029,label = f'Suscetíveis {file[9:-4]}')
    plt.plot(infect[0],infect[-3]/2029,label = f'Hospitalizados {file[9:-4]}')
plt.legend()
plt.grid()
#plt.ylim(0,0.1)
#plt.xlim(50,300)
plt.show()
```

```python
plt.plot(infect[0],infect[1]/2029,label = 'Suscetíveis')
plt.plot(infect[0],infect[2]/2029,label = 'Expostos')
plt.plot(infect[0],infect[-2]/2029,label = 'Recuperados')
#plt.plot(hist[0],hist[-3]/2029)
#plt.plot(hist[0],hist[-1]/2029)
plt.legend()
plt.grid()
plt.xlim(0,80)
plt.show()
```
