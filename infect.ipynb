```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.loadtxt("./C/output/infect.txt",delimiter= ' ').T
```

```python
plt.plot(np.cumsum(a[0]),a[1]/2027,label = 'Suscetvel')
plt.plot(np.cumsum(a[0]),a[2]/2027,label = 'Expostos')
plt.plot(np.cumsum(a[0]),a[-2]/2027,label = 'Recuperados')
plt.plot(np.cumsum(a[0]),a[-3]/2027,label = 'Hospitalizados')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(np.cumsum(a[0]),a[2]/2027,label = 'Expostos')
plt.plot(np.cumsum(a[0]),a[3]/2027,label = 'Assintomtico')
plt.plot(np.cumsum(a[0]),a[4]/2027,label = 'Sintomtico')
plt.legend()
plt.grid()
plt.show()
```
